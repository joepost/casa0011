### Task: Fireflies Behaviour Tool
Produce a heat map visualising the results of bivariate parameter testing for the Fireflies NetLogo model.

```python
# load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in data
fireflies = pd.read_csv("fireflies_datacleaned.csv")

print(fireflies.shape)
fireflies.head()
```

```python
# Remove unnecessary columns
ff = fireflies.drop(columns = ['run_number', 'number', 'strategy', 'flashes_to_reset', 'step'])

ff.shape
```

```python
# Clean data to remove runs that didn't finish
ff_cln = ff[ff.synch_time > 0]

ff_cln.shape
```

```python
# Calculate mean per parameter pair 
ff_counts = ff_cln.groupby(["cycle_length", "flash_length"], as_index=False)["synch_time"].count()
ff_counts.head()
```

```python
# Calculate mean per parameter pair 
ff_means = round(ff_cln.groupby(["cycle_length", "flash_length"], 
                                as_index = False     # important to add if wanting output in format dataframe
                               )["synch_time"].mean())
ff_means
```

```python
# Calculate standard deviation
ff_sd = round(ff_cln.groupby(["cycle_length", "flash_length"], 
                                as_index = False     # important to add if wanting output in format dataframe
                               )["synch_time"].std())
ff_sd
```

```python
# Pivot into heatmap format
ff_heatmap = ff_means.pivot(index='cycle_length', columns='flash_length', values='synch_time')
ff_heatmap_sd = ff_sd.pivot(index='cycle_length', columns='flash_length', values='synch_time')
```

```python
# Heatmap of mean values

plt.figure(figsize=(10,8))

h = sns.heatmap(ff_heatmap, cmap="viridis")
h.set_facecolor("lightgrey")
h.set(xlabel = "Flash Length", ylabel = "Cycle Length", title = "Mean time to synchronisation")

plt.savefig('mean-time-to-synchronisation.png')
plt.show()
```

```python
# Heatmap of standard deviations

plt.figure(figsize=(10,8))

h = sns.heatmap(ff_heatmap_sd, cmap="viridis")
h.set_facecolor("lightgrey")
h.set(xlabel = "Flash Length", ylabel = "Cycle Length", title = "Standard deviation of time to synchronisation")

plt.savefig('sd-time-to-synchronisation.png')
plt.show()
```
